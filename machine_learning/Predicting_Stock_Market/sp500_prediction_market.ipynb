Train our data with machine learning

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=200, min_samples_split=100, random_state=1)

train = sp500.iloc[:-100]
test = sp500.iloc[-100:]
predictors = ['Open', 'High', 'Low', 'Close', 'Volume']

model.fit(train[predictors], train['Target'])

# min_sample_split az oversampling jologiri mikonad
# dar bala ma predictors tarif kardim yani model ma az in column ha baraye pishbini estefade konad dar real world projects in
# nahve kar kardan bishtar javab midahad ta inkeaz tamami column ha baraye pishbini estefade konim
###
# dar fit kardane model ma 2 parametr tarif kardim ke parametre aval train mikonad model ra va parametre 2vom hadaf az train ra moshakhas mikonad
```

```python
# dar in masale ma az precision_score estefade mikonim chon bar rooye yek meyare moshakhas darim kar mikonim, `target`, bekharim(1) ya na(0)
from sklearn.metrics import precision_score

preds = model.predict(test[predictors])
```

```python
precision_score(test['Target'], preds)
```

building a backtesting system

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train['Target'])
    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index=test.index, name='Predictions')
    combined = pd.concat([test['Target'], preds], axis=1)
    return combined

# haman proce train kardane modeleman ast dar yek tabe
```

```python
# dar in tabe ma 2500 ra beonvane meghdar row avalie baraye train model tarif kardim, taghriban har sal 250 rooze trade darad
# ke 10 sale an mishavad 2500 row yani ma 10 sale aval ra baraye train modeleman tarif mikonim va 11th sal ra predict mikonim
# sepas 11 sal ra baraye train estefade mikonim va 12th sal ra predict mikonim sepas 12 sal ra baraye train estefade mikonim va 13th sal ra predict mikonim va...
def backtest(data, model, predictors, start=2500, step=250):
    all_predictions = []
    
    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)
    return pd.concat(all_predictions)
```

```python
predictions = backtest(sp500, model, predictors)
```

Adding aditional predictors to our model

```python
# horizons dar vaghe rooz haye akhir hastand ke mikhahim bar asase anha pishbini konim
horizons = [2, 5, 60, 250, 1000]
new_predictors = []

for horizon in horizons:
    rolling_averages = sp500.rolling(horizon).mean()
    
    ratio_column = f'Close_Ratio_{horizon}'
    sp500[ratio_column] = sp500['Close'] / rolling_averages['Close']
    
    trend_column = f'Trend{horizon}'
    sp500[trend_column] = sp500.shift(1).rolling(horizon).sum()["Target"]
    
    new_predictors += [ratio_column, trend_column]
```

#### Improving our model

```python
model = RandomForestClassifier(n_estimators=200, min_samples_split=50, random_state=1)
```

```python
# ma dobare tabe predict ra minevsiim vali ba kami taghirat, ma mikhahim modeleman nahve tasmim giri rajebe 0, 1 ra taghir dahad
# yani ehtemali be ma bedahad ke bishtar 1 ast ya 0
def predict(train, test, predictors, model):
    model.fit(train[predictors], train['Target'])
    preds = model.predict_proba(test[predictors])[:,1]
    preds[preds >= 0.6] = 1
    preds[preds < 0.6] = 0
    preds = pd.Series(preds, index=test.index, name='Predictions')
    combined = pd.concat([test['Target'], preds], axis=1)
    return combined
```

```python
predictions = backtest(sp500, model, new_predictors)
```

```python
predictions['Predictions'].value_counts()
```

```python
precision_score(predictions['Target'], predictions['Predictions'])
```
