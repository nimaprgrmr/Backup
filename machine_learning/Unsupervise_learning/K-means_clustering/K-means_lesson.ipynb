```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
# baraye dorost kardane data haye masnooyi az in khosusiyate sklearn estefade mikonim
```

```python
data = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.8, random_state=101)
# dar inja meghdare cluster_std = (baraye tayin kardane parakandegie dade ha estefade mishavad) be onvane mesal
# harche adad balatar bashad parakandegi bishtar va harche adad kamtar bashad tarakom har cluster bishtar
```

```python
plt.scatter(data[0][:,0], data[0][:,1] , c=data[1], cmap='rainbow')
```

```python
# amoozeshe model 
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(10,5))

ax1.set_title("K_means")
ax1.scatter(data[0][:,0], data[0][:,1], c=kmeans.labels_, cmap='rainbow')

ax2.set_title("Original")
ax2.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='rainbow')
```
